Estadística de la longitud de los transcriptomas extraídos.

```python
import mysql.connector
import pandas as pd

conn = mysql.connector.connect(
   host="localhost",
   database="tesis2",
   user="root",
   passwd=""
)

df = pd.read_sql_query("SELECT id_especie, LENGTH(secuencia) longitud, flg_pct FROM secuencias WHERE flg_seleccionado = 1", conn)
df.hist("longitud")

conn.close()
```
